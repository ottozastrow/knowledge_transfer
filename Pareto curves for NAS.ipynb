```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
target = 0.2
accs = np.arange(0.90, 0.96, 0.001)
lats = np.arange(0.001, 0.2, 0.0005)
```

```python
def funct(acc, lat, w):
    '''function to optimize during architecture search
    Args:
        acc (float): val_miou
        lat (float): latency in s
        w (float): weight '''
    return acc * ((lat/target) ** w)
```

```python
def drawline(at):
    optimal_x = []
    optimal_y = []
    for x in range(len(lats)):
        for y in range(len(accs)):
            a = funct(accs[y], lats[x], -0.01)
            if(abs(a-at)<0.0001):
                optimal_x.append(lats[x])
                optimal_y.append(accs[y])
    plt.plot(optimal_x, optimal_y, 'blue')

```

```python
ats = np.arange(0.8, 1.2, 0.01)
for i in range(len(ats)):
    drawline(ats[i])
plt.show()
```
